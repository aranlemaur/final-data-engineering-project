```
# Data ingestion
file201901 = "/FileStore/2019_01.csv"
file201902 = "/FileStore/2019_02.csv"
file201903 = "/FileStore/2019_03.csv"
fileAirports = "/FileStore/airports.csv"
fileCarriers = "/FileStore/carriers.csv"
fileAirplanes = "/FileStore/plane_data.csv"
```

```
#Creating DFs for every file
fileLocation = "/FileStore/2019_01.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

DF201901=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)

fileLocation = "/FileStore/2019_02.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

DF201902=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)

fileLocation = "/FileStore/2019_03.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

DF201903=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)

fileLocation = "/FileStore/airports.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

airportsDF=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)

fileLocation = "/FileStore/carriers.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

carriersDF=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)

fileLocation = "/FileStore/plane_data.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

planedataDF=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)
```

```
#0. Join datasets - DF201901, DF201902, DF201903
DFJoined = DF201901.union(DF201902).union(DF201903)
display(DFJoined)
```

```
from pyspark.sql.functions import col
DFJoinedDelays = DFJoined.filter(col("ARR_DELAY") > 0).count()
print("Number of delayed flights:")
display(DFJoinedDelays)
DFJoinedBeforeSchedule = DFJoined.filter(col("ARR_DELAY") < 0).count()
print("\nNumber of flights arriving before schedule:")
display(DFJoinedBeforeSchedule)
DFJoinedNulls = DFJoined.filter(col("ARR_DELAY").isNull()).count()
print("\nNumber of null flights:")
display(DFJoinedNulls)
```

```
# When is the best time of the month/day of week/time of year to fly to minimise delays?

# DATASETS to use:
#1.	2019-01.csv
#2.	2019-02.csv
#3.	2019-03.csv

#Steps:
#0. Join datasets
#1. Choose relevant columns per interval of time
#2. Produce graphs
```

```
#1. Choose relevant columns per interval of time.
#1.1 When is the best day of week to fly to minimise delays?
from pyspark.sql.functions import col
DFJoinedDay = DFJoined.select(col("DAY_OF_WEEK"), col("ARR_DELAY"))
print("Delayed flights per day of week")
display(DFJoinedDay)
```

```
#1.2. When is the best time of the month to fly to minimise delays?
from pyspark.sql.functions import col
DFJoinedMonth = DFJoined.select(col("DAY_OF_MONTH"), col("ARR_DELAY"))
print("Delayed flights per day of month")
display(DFJoinedMonth)
```

```
#1.3. When is the best time of year to fly to minimise delays?
#1.3.1. Only 3 first months of 2019
from pyspark.sql.functions import col
DFJoinedYear = DFJoined.select(col("MONTH"), col("ARR_DELAY"))
print("Delayed flights per month")
display(DFJoinedYear)
```

```
#1.4. When is the best time of the day to fly to minimise delays?
from pyspark.sql.functions import col
DFJoinedTimeIntervals = DFJoined.select(col("CRS_DEP_TIME"), col("ARR_DELAY"))

#Time interval => midnight to 8am
DFJoinedTimeIntervals1 = DFJoinedTimeIntervals.filter((DFJoinedTimeIntervals["CRS_DEP_TIME"] >0) & (DFJoinedTimeIntervals["CRS_DEP_TIME"] <= 800))
display(DFJoinedTimeIntervals1)
```

```
#Time interval => 8am to noon
DFJoinedTimeIntervals2 = DFJoinedTimeIntervals.filter((DFJoinedTimeIntervals["CRS_DEP_TIME"] >800) & (DFJoinedTimeIntervals["CRS_DEP_TIME"] <=1200))
display(DFJoinedTimeIntervals2)
```

```
#Time interval => noon to 5pm
DFJoinedTimeIntervals3 = DFJoinedTimeIntervals.filter((DFJoinedTimeIntervals["CRS_DEP_TIME"] >1200) & (DFJoinedTimeIntervals["CRS_DEP_TIME"] <=1700))
display(DFJoinedTimeIntervals3)
```

```
#Time interval => 5pm to 9pm
DFJoinedTimeIntervals4 = DFJoinedTimeIntervals.filter((DFJoinedTimeIntervals["CRS_DEP_TIME"] >1700) & (DFJoinedTimeIntervals["CRS_DEP_TIME"] <=2100))
display(DFJoinedTimeIntervals4)
```

```
#Time interval => 9pm to midnight
DFJoinedTimeIntervals4 = DFJoinedTimeIntervals.filter((DFJoinedTimeIntervals["CRS_DEP_TIME"] >2100) & (DFJoinedTimeIntervals["CRS_DEP_TIME"] <=2400))
display(DFJoinedTimeIntervals4)
```

```
# Do older planes suffer more delays?
# DATASETS to use:
#1.	2019-01.csv
#2.	2019-02.csv
#3.	2019-03.csv
#4. plane_data.csv

#Steps:
#1. Join files - get DFJoined dataset, and append planedataDF. Columns shared: TAIL_NUM and tailnum. Rename "year" column to avoid ambiguity
#2. Choose relevant columns
#3. Produce graphs
```

```
#1. Join files - get DFJoined dataset, and append planedataDF. Columns shared: TAIL_NUM and tailnum. Rename "year" column to avoid ambiguity
planedataDF = planedataDF.withColumnRenamed("year","manufacture_year")
datesPlanesDF = DFJoined.join(planedataDF).where(DFJoined["TAIL_NUM"] == planedataDF["tailnum"])
display(datesPlanesDF)
```

```
from pyspark.sql.functions import col
#2. Choose relevant columns
planeAgeDF = datesPlanesDF.select(col("ARR_DELAY"), col("manufacture_year"))

#3. Produce graph
display(planeAgeDF)
```

```
# Explore what happens in 1986 vs 1987
planeAgeDF86 = planeAgeDF.where(planeAgeDF.manufacture_year=="1986")
display(planeAgeDF86)
print("Number of planes flying in 1986:")
planeAgeDF86 = planeAgeDF.where(planeAgeDF.manufacture_year=="1986").count()
display(planeAgeDF86)

planeAgeDF87 = planeAgeDF.where(planeAgeDF.manufacture_year=="1987")
display(planeAgeDF87)
print("\nNumber of planes flying in 1987:")
planeAgeDF87 = planeAgeDF.where(planeAgeDF.manufacture_year=="1987").count()
display(planeAgeDF87)
```

```
# How does the number of people flying between different locations change over time?
# DATASET to use:
#1.	2019-01.csv -> DF201901
#2.	2019-02.csv -> DF201902
#3.	2019-03.csv -> DF201903
#4. airports.csv
#(DFJoined)

#Steps:
#0. Join tables on key iata == origin/dest
#1. Identify busiest airports in ORIGIN and DEST
#2. Focus on these columns: MONTH, ORIGIN, DEST
#3. Produce tables
```

```
#0. Join tables (ORIGIN == IATA)
airportsAndMonth1DF = DF201901.join(airportsDF).where(DF201901["ORIGIN"] == airportsDF["iata"])
airportsAndMonth2DF = DF201902.join(airportsDF).where(DF201902["ORIGIN"] == airportsDF["iata"])
airportsAndMonth3DF = DF201903.join(airportsDF).where(DF201903["ORIGIN"] == airportsDF["iata"])
```

```
#1. Identify busiest origin airports in January
busiestAirports1 = airportsAndMonth1DF.groupBy("ORIGIN", "airport").count()
display(busiestAirports1)
```

```
#1. Identify busiest origin airports in February
busiestAirports2 = airportsAndMonth2DF.groupBy("ORIGIN", "airport").count()
display(busiestAirports2)
```

```
#1. Identify busiest origin airports in March
busiestAirports3 = airportsAndMonth3DF.groupBy("ORIGIN", "airport").count()
display(busiestAirports3)
```

```
#0. Join tables (DEST == IATA)
airportsAndMonth1DESTDF = DF201901.join(airportsDF).where(DF201901["DEST"] == airportsDF["iata"])
airportsAndMonth2DESTDF = DF201902.join(airportsDF).where(DF201902["DEST"] == airportsDF["iata"])
airportsAndMonth3DESTDF = DF201903.join(airportsDF).where(DF201903["DEST"] == airportsDF["iata"])
```

```
#1. Identify busiest destination airports in January
busiestAirportsDEST1 = airportsAndMonth1DESTDF.groupBy("DEST", "airport").count()
display(busiestAirportsDEST1)
```

```
#1. Identify busiest destination airports in February
busiestAirportsDEST2 = airportsAndMonth2DESTDF.groupBy("DEST", "airport").count()
display(busiestAirportsDEST2)
```

```
#1. Identify busiest destination airports in March
busiestAirportsDEST3 = airportsAndMonth3DESTDF.groupBy("DEST", "airport").count()
display(busiestAirportsDEST3)
```
